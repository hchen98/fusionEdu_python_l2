## Function

Q: what is a function?

function2.png

Let's say we want to calculate the area of a triangle. We can do it by creating a function:

$\Large A = \frac{1}{2} * b * h$

such that:
- $\Large A$: area, (aka LHS) is the output of a func.
- $\Large \frac{1}{2} * b * h$: the body of a func (aka RHS).
- $\Large b \ \text{and} \ h$: the inputs of a func.

*LHS: Left Hand Side

*RHS: Right Hand Side

function3.png

### Example

```python
def triangle_area(b, h):
  area = 1/2 * b * h
  return area


print(triangle_area(43.9, 99.03))
```

### Multiple functions

Just like variables, we can create multiple functions.

```python
# create three different functions
def func1():
  pass


def func2():
  pass


def func3():
  pass

# calling those three different functions
func1()
func2()
func3()
```

However, sometimes, we may want to create two same function naming (more detail in OOP section!).

Let's do an experiment.

```python
# I create two func with the same naming
def func1():
  print("Hooray!")

def func1():
  print("Boo!")


# Q: what will the output be?
# calling the func
# func1()
```

```python
# for the same experiemnt, we can do it on variable
number = 10
number = 90

# print(number)
```

What just happened?

Well, that is called ***Overriding*** in programming where we replace an existing variable's value with a new value! 

*Note that we keep the variable naming the same!

## Import

Sometimes, when we have very long code, we break it down into different func. However, when we have so many func, we break it into multiple files.

For example, a file may contain millions lines of code. A file may contain multiple functions. Therefore, we need to come up with a way to orgnize our code so that other programmer will know where to look at it!

import2.png

We do it through `import`!

### Simple function

```python
# test.py
def check():
  return "Hooray!"


#################################


# main.py
from test import check

print(check())
```

import1.png

### Multiple functions

```python
# test.py
def check1():
  return "This is check 1 function!"

def check2():
  return "This is check 2 function!"


#################################


# main.py
from test import check1, check2

print(check1())
print(check2())
```

import3.png

Q: what happen if we have one million different func in a single file?

```python
# test.py
def check1():
  return "This is check 1 function!"

def check2():
  return "This is check 2 function!"

def check3():
  return "This is check 3 function!"

def check4():
  return "This is check 4 function!"


#################################


# main.py
from test import *

print(check1())
print(check2())
print(check3())
print(check4())
```
