# String Formatting

In Python, we know how to print a message on a console through `print()` function. In Level 2, we are going to go depth with that!

```python
print("Hooray!")
```

Q: what happen if we want to print the value from a variable along with the text?

```python
name = "Henry"

print("My name is ")
print(name)
```

## %

One simply way to fix that is through Python's built-in operation `%`.

```python
name = "Henry"

print("My name is %s" % name)
# print(name)
```

### Disadvantage

Lack of readability when we want to print multiple variables! 

```python
weather = 99.2
name = "Henry"
age = 24
height = 5.3
weight = 900
student = False

print("Today's weather is %s, %s went to shopping with his friend. %s 's info is age: %s, height: %s, weight: %s, and his is a student: %s" %(weather, name, name, age, height, weight, student))
```

## `format()`

Alternative way to replace `%` is `format()` function.

```python
name = "Henry"

print("My name is {}".format(name))
```

```python
name = "Henry"
age = 24

print("My name is {} and {} years old.".format(name, age))
```

image.png

### Disadvantage

The problem with `format()` is that it may end up too lengthy. So it didn't solve the problem that we faced in `%` operator.

```python
weather = 99.2
name = "Henry"
age = 24
height = 5.3
weight = 900
student = False

print("Today's weather is {}, {} went to shopping with his friend. {} 's info is age: {}, height: {}, weight: {}, and his is a student: {}".format(weather, name, name, age, height, weight, student))
```

Q: can we do better with it? 🤔

## f string

```python
name = "Henry"

print(f"My name is {name}")
```

```python
weather = 99.2
name = "Henry"
age = 24
height = 5.3
weight = 900
student = False

print(f"Today's weather is {weather}, {name} went to shopping with his friend. {name} 's info is age: {age}, height: {height}, weight: {weight}, and his is a student: {student}")
```

```python
weather = 99.2
name = "Henry"
age = 24
height = 5.3
weight = 900
student = False

print(f'Today's weather is {weather}, {name} went to shopping with his friend. {name} 's info is age: {age}, height: {height}, weight: {weight}, and his is a student: {student}')
# print(f'Todays weather is {weather}, {name} went to shopping with his friend. {name} s info is age: {age}, height: {height}, weight: {weight}, and his is a student: {student}')
```

## join()

Sometimes, we may want to have special characters in between the text charaters.

e.g., put dash in between the word Hooray.

Expected output: H-o-o-r-a-y

```python
word1 = "Hooray"

print("-".join(word1))
```

```python
word1 = "Hooray"
word2 = "_"

print(word2.join(word1))
```

Q: we can do it with other data types?

```python
word1 = "Hooray"

print(9.join(word1))
```

As we can see, `join()` function only works with string data type!

Q: does it work with different data structure?

e.g., tuple, set, list, and dictionary?

```python
words = ["aa", "bb", "cc", "dd"]

print("_".join(words))
```

```python
words = {
    "aa": "a",
    "bb": "b",
    "cc": "c",
    "dd": "d"
    }

print("_".join(words))
```
